```python
df= pd.read_json("files/dados_hospedagem.json")
```

```python
#extrair dados usando json normalize
df = pd.json_normalize(df["info_moveis"])
df
#todos os dados estão dentro de uma lista []
```

```python
#explode colunas 0 até 2
df = df.explode(columns[3:])
df
```

### Conversão dados numéricos

```python
import numpy as np
```

```python
#converter de 
df["max_hospedes"] = df["max_hospedes"].astype(np.int64)
```

```python
df.dtypes
```

```python
columns_numerics = ["quantidade_banheiros", "quantidade_quartos", "quantidade_camas"]
```

```python
df[columns_numerics] = df[columns_numerics].astype(np.int64)
```

```python
df.dtypes
```

```python
df["avaliacao_geral"]
```

```python
#convertendo para float
df["avaliacao_geral"] = df["avaliacao_geral"].astype(np.float64)
```

```python
df.dtypes
```

```python
df["preco"]
```

```python
#df["preco"] = df["preco"].astype(np.float64)
#String "$" não pode ser convertida para float
```

```python
df["preco"] = df["preco"].apply(lambda x: x.replace("$","").replace(",", "").strip())
#funão lambda = função anônima para rápidas expressões
#função replace = substitui uma uma string por outra, neste caso de funções anônimas nesse caso, de "$" para ""
#função strip = remove espaços em branco, ou vazios
```

```python
df.dtypes
```

```python
df["preco"] = df["preco"].astype(np.float64)
```

```python
df.dtypes
```

```python
#Erro por conta do caractere '$'
df[["taxa_deposito", "taxa_limpeza"]] = df[["taxa_deposito", "taxa_limpeza"]].astype(np.float64)
```

```python
#applymap porque tem 2 registros a serem modificados
df[["taxa_deposito", "taxa_limpeza"]] = df[["taxa_deposito", "taxa_limpeza"]].map(lambda x: x.replace("$","").replace(",", "").strip())
```

```python
df[["taxa_deposito", "taxa_limpeza"]] = df[["taxa_deposito", "taxa_limpeza"]].astype(np.float64)
```

```python
df.dtypes
```

## Conversão dados textuais

```python
df["descricao_local"]
```

```python
df["descicao_local"] = df["descricao_local"].str.lower()
```

```python
df["descricao_local"][300]
```

```python
df["descricao_local"] = df ["descricao_local"].str.replace("[^a-zA-Z0-9\-\']", " ", regex=True)
```

```python
df["descricao_local"][300]
```

```python
df["descricao_local"]=df["descricao_local"].str.replace("(?<!\w)-(?!\w)", " ", regex=True)
```

```python
df["descricao_local"]
```

```python
df.head(4)
```

```python
#cria um tolken de palavras que pode ser usado futuramente em uma núvem de palavras
df["descricao_local"] = df["descricao_local"].str.split()
```

```python
df.head(3)
```

```python
#Remove caracteres especiais, neste caso queremos tirar as chaves {}
df["comodidades"] = df["comodidades"].str.replace(r"[{}\”]", "", regex=True)
```

```python
df["comodidades"]
```

```python
#transformando em lista
df["comodidades"] = df["comodidades"].str.split(",")
```
